```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
year = np.array((5, 10, 4, 10, 20))
salary = np.array((10, 20, 6, 70, 80))
print(year)
print(salary)
```

```python
mean_x= np.mean(year)
mean_y = np.mean(salary)
print(mean_x)
print(mean_y)
```

```python
slope_upper = 0.0
slope_lower = 0.0
for i,j in zip(year, salary):
    slope_upper = (i-mean_x)*(j-mean_y)
    slope_lower = (i-mean_x)**2
slope = slope_upper / slope_lower
print(slope)
```

```python
intercept = mean_y - (slope * mean_x)
print(intercept)
```

```python
print("y =", end=" ")
print(slope,"x +",intercept)

eq = slope * year + intercept
print(eq)
```

```python
fig = plt.figure(figsize = (10,5))
#Scatter Plot The blue dots
# plt.scatter(year, salary, color = 'blue', label = 'Data Points')
plt.plot(year, eq, color = 'red', label = "Linear Regression Equation")
plt.xlabel('Years')
plt.ylabel('Salary')
#legend to display that corner thing dabba in label
plt.legend()
plt.show()
```
