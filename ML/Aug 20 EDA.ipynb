```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'Name' : ['Alice', 'Bob', 'Charlie', 'David', 'Emily','Rish','Vijay','Sid','Shukla','Sam'],
    'Age':[25 ,30, 20, 28, 24,10,20,30,40,50],
    'Height': [165,170,160,175, 180, 120, 156,140,200,150],
    'Weight':[80 ,87, 55, 58, 60,70,50,40,66,56],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female','Male', 'Male', 'Male', 'Female','Male'],
    'Math_Score':[85, 92, 78, 88, 76,10,100,21,40,2],
    'Science_Score':[90, 90, 85, 92, 88,10,99,100,2,3],
    'Passed_Exam':['Yes', 'Yes', 'No', 'Yes', 'No','No','Yes', 'Yes', 'No','No']
}

df = pd.DataFrame(data)
print(df)
```

```python
#Step 2 :Univariate Analysis : Numerical Data

plt.figure(figsize=(10,4))
# plt.subplot(1,2,1)
sns.histplot(df['Age'],kde=True,color='blue')
plt.title('Age Distribution')
# plt.show()

plt.figure(figsize=(10,4))
# plt.subplot(1,2,1)
sns.histplot(df['Age'],kde=True)
plt.title('Age Distribution')



# plt.figure(figsize=(10,4))
# plt.subplot(1,2,2)
# sns.histplot(df['Weight'],kde=True,color='orange')
# plt.title('Age Distribution')
# plt.tight_layout()
# plt.show()
```

```python
# Bivariate Analysis

sns.scatterplot(x='Age',y='Weight',hue='Gender',data=df)
plt.title('Scatter Plot ')
plt.show()

sns.boxplot(y='Age',hue='Gender',data=df)
plt.title('BOXPLOT')
plt.show()

sns.histplot(x='Age',data=df,kde=True)
plt.title('HISTPLOT')
plt.show()

sns.countplot(x='Gender',hue='Gender',data=df)
plt.title('COUNTPLOT')
plt.show()
```

```python
# Step 3:  Categorical Data

# plt.figure(figsize=(10,4))
# plt.subplot(1,2,1)
sns.countplot(x='Gender',hue='Gender',data=df)
plt.title("Gender Count(categorical data)")

# plt.subplot(1,2,2)
# sns.countplot(x='Passed_Exam',hue='Gender',data=df,palette='Set2')
# plt.title("Passed Exam Count(categorical data)")

# plt.tight_layout()
# plt.show()
```

```python
#Step 4 Bivariate Analysis-Numerical Data
sns.scatterplot(x='Age',y='Weight',hue='Gender',data=df)
plt.title("Scatter Plot of Age vs Weight")
plt.show()
```

```python
# Step 5: Oultier Detection
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.boxplot(data=df['Science_Score'],color='red')
plt.title("Outlier in Science Score(box-plot)")

plt.subplot(1,2,2)
sns.boxplot(data=df['Height'],color='green')
plt.title("Outlier in Height(box-plot)")
```
