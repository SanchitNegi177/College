```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
```

```python
df.describe()
```

```python
print("Data Types",df.dtypes)
```

```python
# Fill missing values in numeric columns with the mean

numeric_cols = df.select_dtypes(include=['int64', 'float64'])
df[numeric_cols.columns] = numeric_cols.fillna(numeric_cols.mean())


non_numeric_cols = df.select_dtypes(include=['object'])
for col in non_numeric_cols.columns:
    mode_value = df[col].mode()[0]  # Get the mode (most frequent value)
    df[col].fillna(mode_value, inplace=True)

print(df.isnull().sum())
```

```python
sns.boxplot(df['Engine HP'])
plt.title('Box Plot for MSRP')
plt.figure(figsize=(100, 100))
plt.ylabel('Engine HP')
plt.show()
```

```python
plt.boxplot(df['Engine Cylinders'])
plt.title('Box Plot for Engine Cylinders')
plt.ylabel('Engine HP')
plt.show()

plt.boxplot(df['Engine HP'])
plt.title('Box Plot for Engine HP')
plt.ylabel('Engine HP')
plt.show()

plt.boxplot(df['MSRP'])
plt.title('Box Plot for MSRP')
plt.ylabel('Engine HP')
plt.show()

```
